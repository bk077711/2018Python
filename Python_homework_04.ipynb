1) 파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오.

>with ~ as 문을 사용하였고 split('\n')으로 개행문자 기준으로 리스트를 만든 뒤 앞의 단어의 오름차순으로 리스트를  
정렬한 후 '\n'.join으로 다시 문자열로 합친 뒤 s1.txt에 입력함
***

```python
with open('s.txt', 'r') as f:
    s = f.read()
    dic = {}
for line in s.split('\n'):
    i = (line.split(' '))
    dic[i[0]] = i[1]
s = sorted(dic.items(), key = lambda x : x[1])

with open('s2.txt', 'w') as f:
    for i in s:
        f.write(i[0] + ' ' + i[1] + '\n')
with open('s2.txt', 'r') as f:
          s = f.read()
          print(s)
```

>with ~ as 문으로 s.txt를 읽고 이를 s에 저장한다. 그리고 빈 사전 자료형 dic을 선언하고

그리고 line을 공백 문자를 기준으로 나눠서 리스트형 i에 저장하고 i[0]은 키값 i[1]은 벨류값으로 사전에 저장한다.  
그리고 sorted()함수로 벨류값 기준 오름차순으로 정렬해서 리스트형 s에 저장한다. 그리고 리스트 s를  
"키값 벨류값"으로 파일로 출력하면 s2.txt가 된다.
***

```python
with open('s.txt', 'r') as f:
    s = f.read()
s = s.split()

with open('s3.txt', 'w') as f:
    for i in range(len(s)):
        f.write(s[i])
        if i % 3 == 2:
            f.write('\n')
        else:
            f.write(' ')
with open('s3.txt', 'r') as f:
    s = f.read()
print(s)
```

### [3번 문제]
range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

#### 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.  
range(stop)  
range(start, stop[, step])  

#### 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.  
#### 3) frange 사용 예  
+ 3-1) frange(0.5)  
[0.0, 0.1, 0.2, 0.3, 0.4]
+ 3-2) frange(1.0, 2.0)  
[1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
+ 3-3) frange(2.2, 4.0, 0.5)  
[2.2, 2.7, 3.2, 3.7]

```python
def frange(one, two = 'x', three = 'x'):
    l = []
    if two == 'x' and three == 'x' :
        stop = one
        start = 0.0
        step = 0.1
    elif two != 'x' and three == 'x':
        start = one
        stop = two
        step = 0.1
    elif two != 'x' and three != 'x':
        start = one
        stop = two
        step = three
    while 1:
        start
        if start < stop:
            l.append(round(start,2))
        else:
            break
        start = start + step    
    return l

print(frange(0.5))
print(frange(1.0,2.0))
print(frange(2.2, 4.0, 0.5))
```

***
우선 frange의 형태는 3가지가 있다.
+ 1번째 stop만 주어진 경우
+ 2번째 start와 stop이 주어진 경우
+ 3번째 세 개의 인자가 모두 주어진 경우
세 가지 경우의 공통점은 stop인자는 무조건 입력받는다는 것이다.  
그렇기에 디폴트 매개변수를 이용해서 인자를 입력 받은 뒤에  디폴트 값인 'x'의 개 수에 따라 start와 stop과 step을 설정했다.  
'x' 인자가 2개 인 경우 stop만 주어졌기에 start와 step 은 0.0, 0.1이 된다.  
'x' 인자가 1개 인 경우 start와 stop이 주어졌고 step은 0.1이 된다.  
'x' 인자가 없을 경우는 각각 입력 받은 값이 된다.  

start stop step 값이 정해진 후에는
start와 step을 비교한 뒤에 start < step일 경우에는 리스트 l에 start값을 저장하고
start 값에 step을 더한 뒤 while문이 처음붜 반복된다. 그러다가 start값이 step값보다 커지면 break 되고 리스트 l이 반환된다.

### [4번 문제]
가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오  

#### sum() 사용 예  
+ sum()  
 0  
+ sum(1, 2)  
 3  
+ sum(1, 2, 3, 4, 5)  
 15  
+ sum(1, 5, 7, 2, -10)  
 5  

```python
def sum(*var):
    result = 0
    for i in var:
        result = i + result
    return result
print(sum())
print(sum(1, 2))
print(sum(1, 2, 3, 4, 5))
print(sum(1, 5, 7, 2, -10))
```

***
sum()함수는 사용자로 부터 가변인수를 인자로 받는다 그러면 그 숫자들을 모두 더한 뒤에 출력한다.  
이를 구현하기 위해서 def sum(\*var)는 모든 인자들을 리스트에 저장한다.  
그러면 for i in var 문을 해서 var의 모든 리스트 원소들을 i에 대입한다. 그리고 result에 더한 뒤 result 값을 반환하면  
sum()함수가 구현된다.

+ 1) 다음에 제시되는 함수들을 모두 이용해야 한다.  
>split  
>map  
>join  
+ 2) myinitial() 함수 사용 예  
>myinitial("as soon as possible")  
>ASAP  

```python
def myinitial(s):
    s = s.split('\n')
    l = []
    for word in s:
        word = word.split(' ')
        for i in word:
            l.append(i[0])
    l = list(map(ord,l))
    for i in range(len(l)):
        l[i] = l[i] - 32
    l = list(map(chr,l))
    result = ''.join(l)
    return result

s = '''asdf asddffsf
ffff'''
print(myinitial(s))
print(myinitial("as soon as possible"))
```

***

그리고 분류한 값들의 첫번째 문자를 l에 append를 해서 저장한 뒤에 리스트 l의 요소들을

를 해서 아스키코드값으로 바꾸고 -32를 해서 대문자로 변경한뒤 다시 map을 이용해서 아스키코드값을 문자로 바꾼다.  
그리고 ''.join()을 해서 하나의 문자열로 합친 뒤 반환하면 각 단어들의 첫글자가 대문자로 출력되서 나간다.

```python
def myfact(n):
    if n < 0:
        return
    elif n == 0:
        return 1
    elif n <= 1:
        return 1
    else :
        return n* myfact(n-1)
x = 1
while 1:
    print('x = {0}, x! = {1}'.format(x,myfact(x)))
    x = x+1
```

그리고 예외처리해야할 경우가 0 입력과 음수 입력인데 0!의 경우 수학적으로 1으로 '정의'된다. 그렇기에 0일 경우에는 1이 반환되게 했다.

```python
def extractText(source): #HTML태그를 제외하고 순수 텍스트만 추출하는 함수
    text = "" #순수 텍스트 식별자
    while(1) :      
        source = source[source.find('<'):] #'<'를 찾는다
        if (source.find("<!--") == 0) : 
                source = source[source.find("-->")+3:] #주석 내용 건너뜀 
        elif (source.find("<script") == 0) :        
                source = source[source.find("</script>")+9:]  #자바스크립트 내용 건너뜀   
        elif (source.find("<style") == 0) :
                source = source[source.find("</style>")+9:] #스타일 내용 건너뜀
        else :       
            source = source[source.find('>')+1:] #'>'를 찾는다
            for ch in source :    #'>' 이후부터 한글자씩 가져온다
                if ch == '<' : break    # 다시 '<' 문자 등장하면 종료
                elif ch == '\t' or ch == '\n' : continue    #탭,개행 문자는 순수 텍스트에 포함x
                else : text += ch   #순수 텍스트 추가     
            if text.endswith(" ") == False : text += " " #공백으로 분리하기 위해 순수 텍스트 추출 후 뒤에 공백문자 추가
        if(source.find('<') == -1):break #'<' 문자가 더 이상 없는 경우 무한루프종료 
    return text 
import string

def replaceAll(text, dic):  ##문자열을 사전에 맵핑된 문자로 대체로 하는 함수
    for i,j in dic.items():
        text = text.replace(i,j,text.count(i)) #문자열에 사전에 키와 같은 문자가 존재하면 키에 맵핑된 value값으로 대체한다.
    return text   

def stripSpecialChar(text): ##특수문자 제거 함수
    keys = All_stopWord # 불용어
    keys = keys + [ch for ch in string.punctuation] #구두문자를 key리스트로 만듬
    keys.append("★") #구두문자 이외의 특수문자를 key리스트에 추가
    keys.append("☆")
    keys.append("※")
    keys.append("nbsp")#html문자
    keys.append("lt")
    keys.append("amp")
    keys.append("quot")
    keys.append("middot")
    keys.append("uarr")
    keys.append("rarr")
    keys.append("darr")
    keys.append("harr")
    
    values = ['' for _ in keys]  #key리스트의 원소 수 만큼 ''문자를 원소를 갖는 value리스트를 만듬
    replaceDic = dict(zip(keys,values)) #key리스트 key로 value리스트를 value값으로 하는 사전 만듬

    return replaceAll(text,replaceDic) #추출한 순수 텍스트에서 사전에 맵핑된 제거할 문자들을 모두 제거
def wordOfFreq(l): #단어의 빈도수를 사전으로 반환하는 함수
    word = []   
    for w in l: #리스트의 각 원소들이
        if (w in word) == False : word.append(w) #word리스트에 포함된 단어가 아니면 단어를 리스트에 추가 (단어중복제거)          
    freg = [l.count(w) for w in l] #리스트에서 단어의 출현 빈도수를 리스트로 만듬
    
    return dict(zip(word,freg)) #단어와 단어의 빈도수로 맵핑된 사전 반환
import requests
import os
import pickle
def saveUrl(url):
    req = requests.get(url)
    source = req.text
    text =  extractText(source)  #순수 텍스트 추출
    text = stripSpecialChar(text) #특수문자 제거
    l = text.split()            #공백단위분리
    data = wordOfFreq(l)
    url = url.lstrip('https://')
    url_html = url.lstrip('http://') + '.html'
    url_dic =  url.lstrip('http://') + '.words_frequency.pickle'
    if os.path.exists(url_html) == False:
        f = open(url_html, 'w')
        f.close()
    if os.path.exists(url_dic) == False:
        f = open(url_dic, 'w')
        f.close()
    with open(url_html, 'wb') as f:
        pickle.dump(source,f)
    with open(url_dic, 'wb') as f:
        pickle.dump(data,f)
url_1 = 'http://cse.koreatech.ac.kr'
url_2 = 'http://www.naver.com'
url_3 = 'http://www.daum.net'
url_4 = 'http://www.koreatech.ac.kr'
url_5 = 'https://github.com'
saveUrl(url_1)
saveUrl(url_2)
saveUrl(url_3)
saveUrl(url_4)
saveUrl(url_5)

def loadUrl(url):
    url = url.lstrip('https://')
    url_dic =  url.lstrip('http://') + '.words_frequency.pickle'
    if os.path.exists(url_dic):
        dicData = {}
        try:
            with open(url_dic, 'rb') as f:
                data = pickle.load(f)
                dicData.update(data)
        except EOFError:
            pass
        return sorted(dicData.items(), key=lambda x:x[1], reverse=True)
def topWord(source):
    word = []
    for i in range(len(source)):
        word.append(source[i][0])
        if i == 2 :
            break
    return word
dic1 = loadUrl(url_1)
dic2 = loadUrl(url_2)
dic3 = loadUrl(url_3)
dic4 = loadUrl(url_4)
dic5 = loadUrl(url_5)
print(topWord(dic1))
print(topWord(dic2))
print(topWord(dic3))
print(topWord(dic4))
print(topWord(dic5))
```

주어진 불용어 사이트에 있는 불용어들을 영어는 리스트 en, 한글은 리스트 kr에 저장한다.  
그리고 리스트 내포를 사용해서 라인 단위, 즉 단어 단위로 저장한 뒤, 그 리스트들의 합을 All_stopWord라 한다.  
그리고 기존에 있었던 특수문자를 제거하는 stripSpecialChar(text)가 All_stopWord에 있는 불용어들도 제거하도록  
keys의 초기값을 All_stopWord로 바꿔준다. 그리고 url을 입력받으면 html과 pickle로 나눠서 저장하는 saveUrl함수를 만든다.  
해당 함수는 기존에 있던 메인부분을 함수 내부로 옮겨주고 필요한 파일이 없으면 생성하고 존재할 경우 pickle모듈의 dump로 파일에  
작성한다.  
그리고 loadUrl과 topWord 함수를 만드는데loadUrl 함수는 피클 파일을 불러와서 그것을 피클 모듈의 함수인 load()를 써서 저장한 뒤에  
파일 입력이 끝나면 에러가 발생하므로 이를 처리하기 위해서 try ~ except 구문을 써서 pass하고 sorted 함수를 써서  
value값 기준으로 내림차순 정렬한 리스트를 반환하였다.  
그리고 topWord 함수에서 가장 value값이 큰 ( 가장 빈도가 잦은 ) 단어 3개를 저장하고 반환한다.  
***

```python
def searchEngine(text):
    stripSpecialChar(text)
    t = text.split('\n')
    s = ' '.join(t)
    t = s.split(' ')
    dic1 = loadUrl(url_1)
    dic2 = loadUrl(url_2)
    dic3 = loadUrl(url_3)
    dic4 = loadUrl(url_4)
    dic5 = loadUrl(url_5)
    url_l = (dict(dic1),dict(dic2), dict(dic3), dict(dic4), dict(dic5))
    url_st = ''
    url_dic = {}
    for i in range(len(url_l)) :
        url_st = ' '.join(url_l[i].keys())
        for word in t:
            cnt = 0
            if word in url_l[i] :
                cnt = cnt + url_l[i][word]
            for line in url_st.split(' '):
                if line.find(word) != -1 :
                    cnt = cnt + 0.5
            if i == 0:
                url_dic.update({ url_1 : cnt})
            if i == 1:
                url_dic.update({ url_2 : cnt})
            if i == 2:
                url_dic.update({ url_3 : cnt})
            if i == 3:
                url_dic.update({ url_4 : cnt})
            if i == 4:
                url_dic.update({ url_5 : cnt})
    item = list(url_dic.items())
    item.sort()
    item.sort(key = lambda x:x[1], reverse = True)
    result = []
    for i in item:
        result.append(i[0])
    return result
text = input()
print(searchEngine(text))
```

***
searchEngine()함수는 사용자로부터 text를 입력받고 그 text를 불용어어를 제거한 순수텍스트로 바꾼다.  
>    stripSpecialChar(text)  
>    t = text.split('\n')  
>    s = ' '.join(t)  
>    t = s.split(' ')  

를 통해서 개행문자와 공백을 없애서 단어별로 t에 저장한다. 그리고 피클로부터 저장된 리스트 dic1~dic5를 사전형식으로 바꿔서  
리스트 url_l 요소로 집어 넣는다. 검색 유사도를 저장할 url_dic을 빈 사전형식으로 선언한다.  
그리고 검색 유사도 기준에 따라서 점수를 매기고 그 점수를 url_dic에 url별로 저장한다. 그리고 url_dic를 list형식으로 바꿔서  
item에 저장하고 key값 기준 오름차순으로 한번, value값 기준 내림차순으로 한번 정렬해서 검색 유사도가 높은 url이 앞에 오고  
나머지는 오름차순으로 오게 정렬이 된다. 그리고 (url, 검색유사도)에서 url만 따로 리스트에 저장해서 반환한다.  
이 문제에서 요즘은 검색 유사도 설정인데 사전 자료형 url_l의 keys값들을 각 url 별로 '\n'기준으로 문자열 url_st에  저장한 뒤 공백 문자를 기준으로  
리스트로 바꿔서 단어별로 비교해서 부분일치하면 0.5점을 더하고 완전히 일치하면 1점을 더한다. 완전 일치와 부분일치가 둘다 될 경우  
중복 되게 점수가 오르지만 여러 개의 부분일치 보다는 더 적은 수의 완전일치가 중요하다고 판단 되었기에 중복 되게 하였다.
