### [2번 문제]
+ 다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.

***
+ 다형성이란
> 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 하는 기능으로서,
> 다른 언어에서는 int형인지 double형인지 등에 대해서 설정해야 하지만 파이썬에서는 형을 선언하지 않아도 됨으로
> 다른 언어와 달리 하나의 메소드만으로도 수많은 형식을 처리할 수 있다.
> 코드가 간결해지는 장점이 있다.
***

```python
class Car:
    def name(self):
        print('...')
class K9(Car):
    def name(self):
        print('this is K9')
class Sonata(Car):
    def name(self):
        print('this is Sonata')
class Tank(Car):
    def name(self):
        print('this is Tank')
class Airplane(Car):
    pass
for each in (K9(), Sonata(), Tank(), Airplane()):
    each.name()
```

***
다형성을 알아보기 위해서 부모 클래스인 Car를 만들고 Car를 상속 받는 클래스들을 만들었다.  
Car 클래스는 name이라는 ...을 출력하는 함수를 멤버 함수로 가지고 있고, Car를 상속받는 클래스의 name은  
자기 이름을 출력하는 함수로 name을 재정의해서 사용한다.

### [3번 문제]
다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

+ 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.

 + count: Counter 인스턴스가 지니는 초기 정수 값
 + step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)
> \>>> c = Counter(10)   
> \>>> d = Counter(10, 2)   

+ 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.

> \>>> print(c)   
> [Count (step: 1)] 10   
> \>>> print(d)  
>[Count (step: 2)] 10  

+ 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

> \>>> c.incr()  
> \>>> d.incr()  
> \>>> print(c)  
> [Count (step: 1)] 11  
> \>>> print(d)  
> [Count (step: 2)] 12  

+ 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.

> \>>> c()  
> \>>> d()  
> \>>> print(c)  
> [Count (step: 1)] 12  
> \>>> print(d)   
> [Count (step: 2)] 14   

+ 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

> \>>> c = c + 5  
> \>>> d = d - 5  
> \>>> print(c)  
> [Count (step: 1)] 17  
> \>>> print(d)  
> [Count (step: 2)] 9  

+ 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

> \>>> print(c) > 10  
> True  
> \>>> print(d) > 10  
> False  
> \>>> print(c) < 10  
> False  
> \>>> print(d) < 10  
> True  
> \>>> print(c) == 17  
> True  
> \>>> print(d) != 9  
> False  

```python
class Counter:
    def __init__(self, count=1, step = 1):
        self.count = count
        self.step = step
    def __str__(self):
        return '[Count (step: {0})] {1}'.format(self.count, self.step)
    def incr(self):
        self.count = self.count + self.step
    def __call__(self):
        self.incr()
        
    def __add__(self, other):
        return Counter(self.count + other, self.step)
    def __sub__(self, other):
        return Counter(self.count - other, self.step)
        
    def __gt__(self, other):
        return self.count > other
    def __ge__(self, other):
        return self.count >= other
    
    def __lt__(self, other):
        return self.count < other
    def __le__(self, other):
        return self.count <= other 
    
    def __eq__(self, other):
        return self.count == other
    def __ne__(self, other):
        return self.count != other
    
c = Counter(10)
d = Counter(10, 2)
print(c)
print(d)
c.incr()
d.incr()
print('------incr() 후------')
print(c)
print(d)
print('-----객체 호출 후-----')
c()
d()
print(c)
print(d)
print('------산술연산 후------')
c = c + 5
d = d - 5
print(c)
print(d)
print('------비교연산------')
print('[c > 10] {0}'.format(c>10))
print('[d > 10] {0}'.format(d>10))
print('[c < 10] {0}'.format(c<10))
print('[d < 10] {0}'.format(d<10))
print('[c == 10] {0}'.format(c==10))
print('[d != 10] {0}'.format(d!=10))
```

### [4번 문제]
다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

+ MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

***
함수가 정의 되어있는 순서는 init , str, eliminate_duplicate이다.
+ init
>    def __init__(self, l):  
>        for e in l:  
>            self.append(e)  
>        MySet.eliminate_duplicate(self)  

init함수는 생성자로 리스트를 받으면 그 리스트의 하위 요소들을 하나씩 리스트에 넣는다.  
그리고 중복을 없애는 eliminate_duplicate함수를 호출하여 중복을 없애서 set자료형을 만든다.

+ str
>    def __str__(self):  
>        result = "MySet: {"  
>        for e in self:  
>            result = result + str(e) + " ,"  
>        result = result[0:len(result)-2] + "}"  
>        return result  

str함수는 print(인스턴스 이름) 할시 호출 되는 함수로 Myset의 모든 요소들을 형식에 맞춰서 추가한다.  
Myset의 요소들을 전부 문자열에 추가한 뒤에는 문자열에 ', '가 끝에 있으므로 [0:len(result)-2]을 해서 ', '을 없앤 문자열을  
반환한다.

+ eliminate_duplicate
>    @staticmethod      
>    def eliminate_duplicate(l):  
>        s = []  
>        for e in l:  
>            if e not in s:  
>                s.append(e)  
>        l[:] = []  
>        for e in s:  
>            l.append(e)  

@staticmethod , 스태틱 메소드는 객체와 무관하게 클래스에 존재하는 정적 메소드이다. 특징으로는 self로 인자를 할당하지 않으며,  
인스턴스 객체가 필요하지 않고 클래스이름으로 직접 호출이 가능하다.  
eliminate_duplicate함수는 set 자료형 l을 받고 리스트형 자료형 s를 만든다.  
그리고 l의 하위 요소 중 s에 포함되지 않은 요소들만 s에 추가한다. 그리고 set형 자료형 l을 슬라이싱을 사용해서 빈 set 자료형으로 만든다.  
그리고 s에 요소들을 l에 추가하면 중복이 제거된다.

### [5번 문제]
4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

+ 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
+ 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

> \>>> u = s | t 
> \>>> print(u) 
> MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 

+ 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

> \>>> u = s & t 
> \>>> print(u) 
> MySet: {2, 3} 

+ 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

> \>>> s = MySet([1, 2, 3]) 
> \>>> t = MySet([3, 4, 5]) 
> \>>> u = s - t 
> \>>> print(u) 
> MySet: {1, 2} 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    
    def __or__(self, other):
        l = []
        for i in self:
            l.append(i)
        for i in other:
            if i not in l:
                l.append(i)
        return MySet(l)
    def __and__(self, other):
        l = []
        for i in self:
            if i in other:
                l.append(i)
        return MySet(l)
    def __sub__(self, other):
        l = []
        for i in self:
            if i not in other:
                l.append(i)
        return MySet(l)
       
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    print()
    u = s | t # __or__
    print(u)
    print()
    u = s & t # __and__
    print(u)
    print()
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t # __sub__
    print(u)
```

***
+ 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
>    def __or__(self, other):
>        l = []
>        for i in self:
>            l.append(i)
>        for i in other:
>            if i not in l:
>                l.append(i)
>        return MySet(l)

두 집합의 합집합을 반환하려면 앞의 MySet클래스 인자를 self로 두고 뒤의 MySet클래스 인자를 other로 두었다.  
여기에서 self와 other은 각각 MySet 클래스 객체이며 for in 문으로 추출할 경우 리스트에서 추출하는 것과 동일한 방법으로 추출이 가능하다.  
그렇기에 빈 리스트 l에 self의 요소들을 전부 넣은 뒤, other으 요소들을 l에 있지 않은 요소들만 추가하면 두 집합의 합집합이 만들어진다.  
이를 return MySet(l)을 하여 두 집합의 합집합을 MySet 클래스 객체로 반환하고 이를 출력하면 s와 t의 합집합인 {1,2,3,4,5,6,7,8,9}이 나오고  
s와 t는 그대로 유지된다.  
+ 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
>    def __and__(self, other):
>        l = []
>        for i in self:
>            if i in other:
>                l.append(i)
>        return MySet(l)

&연산을 이용해서 두 집합의 교집합을 반환하기 위해서 \__and__(self, other) 메소드를 만든다. 해당 메소드는 MySet 클래스 객체를 인자로 받는다.
빈 리스트 l을 만들고, for i in self, if i in other: 를 해서 self의 요소들 중 other와 겹치는 원소들만 리스트 l에 추가한다. 그리고
MySet(l)을 해서 클래스 객체를 반환해주면 두 집합의 교집합이 반환한된다.

+ 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
>    def __sub__(self, other):
>        l = []
>        for i in self:
>            if i not in other:
>                l.append(i)
>        return MySet(l)

두 집합의 차집합을 구하기 위해서 \__sub__를 한다. 차집합 A-B는 B와 중복되지 않는 A의 원소들이기에 self의 원소들 중 other에 포함 되지 않는
모든 원소를 리스트 l에 저장하고 MySet(l)을 해서 클래스 객체를 반환하면 차집합이 반환된다.

```python
import requests
import string
class SearchEngine(list):
    import string
    import requests    
    def __init__(self, *Url):
        for i in Url:
            self.append(i)
        self.data = self.getWordsFrequency()
    def addUrl(self, url, *var):
        self.append(url)
    def removeUrl(self, url, *var):
        if url in self:
            self.remove(url)
    def listUrls(self):
        for i in self:
            print(i)
    @staticmethod        
    def extractText(source): #HTML태그를 제외하고 순수 텍스트만 추출하는 함수
        text = "" #순수 텍스트 식별자
        while(1) :      
            source = source[source.find('<'):] #'<'를 찾는다
            if (source.find("<!--") == 0) : 
                source = source[source.find("-->")+3:] #주석 내용 건너뜀 
            elif (source.find("<script") == 0) :        
                source = source[source.find("</script>")+9:]  #자바스크립트 내용 건너뜀   
            elif (source.find("<style") == 0) :
                source = source[source.find("</style>")+9:] #스타일 내용 건너뜀
            else :       
                source = source[source.find('>')+1:] #'>'를 찾는다
                for ch in source :    #'>' 이후부터 한글자씩 가져온다
                    if ch == '<' : break    # 다시 '<' 문자 등장하면 종료
                    elif ch == '\t' or ch == '\n' : continue    #탭,개행 문자는 순수 텍스트에 포함x
                    else : text += ch   #순수 텍스트 추가     
                if text.endswith(" ") == False : text += " " #공백으로 분리하기 위해 순수 텍스트 추출 후 뒤에 공백문자 추가
            if(source.find('<') == -1):break #'<' 문자가 더 이상 없는 경우 무한루프종료 
        return text 
    
    @staticmethod
    def replaceAll(text, dic):  ##문자열을 사전에 맵핑된 문자로 대체로 하는 함수
        for i,j in dic.items():
            text = text.replace(i,j,text.count(i)) #문자열에 사전에 키와 같은 문자가 존재하면 키에 맵핑된 value값으로 대체한다.
        return text   
    
    @staticmethod
    def stripSpecialChar(text): ##특수문자 제거 함수
        keys = All_stopWord # 불용어
        keys = keys + [ch for ch in string.punctuation] #구두문자를 key리스트로 만듬
        keys.append("★") #구두문자 이외의 특수문자를 key리스트에 추가
        keys.append("☆")
        keys.append("※")
        keys.append("nbsp")#html문자
        keys.append("lt")
        keys.append("amp")
        keys.append("quot")
        keys.append("middot")
        keys.append("uarr")
        keys.append("rarr")
        keys.append("darr")
        keys.append("harr")
        
        values = ['' for _ in keys]  #key리스트의 원소 수 만큼 ''문자를 원소를 갖는 value리스트를 만듬
        replaceDic = dict(zip(keys,values)) #key리스트 key로 value리스트를 value값으로 하는 사전 만듬
        return SearchEngine.replaceAll(text,replaceDic) #추출한 순수 텍스트에서 사전에 맵핑된 제거할 문자들을 모두 제거
    @staticmethod
    def wordOfFreq(l): #단어의 빈도수를 사전으로 반환하는 함수
        word = []   
        for w in l: #리스트의 각 원소들이
            if (w in word) == False : word.append(w) #word리스트에 포함된 단어가 아니면 단어를 리스트에 추가 (단어중복제거)          
        freg = [l.count(w) for w in l] #리스트에서 단어의 출현 빈도수를 리스트로 만듬
        return dict(zip(word,freg)) #단어와 단어의 빈도수로 맵핑된 사전 반환       
         
    def getWordsFrequency(self):
        l = []
        for i in self:
            if i != '':
                req = requests.get(i)
                source = req.text
                text =  self.extractText(source)  #순수 텍스트 추출
                text = self.stripSpecialChar(text) #특수문자 제거
                l += text.split()            #공백단위분리
        return self.wordOfFreq(l)

    def getMaxFreqencyWords(self):
        l = []
        data = sorted( self.data.items(), key=lambda x:x[1], reverse=True)
        cnt = 0
        for i in data:
            if cnt <= i[1]:
                l.append(i[0])
                cnt = i[1]
            else :
                return l
    def searchUrlByWord(self, word):
        similar = []
        for i in self :
            l = []
            cnt = 0
            if word in self.data:
                cnt = cnt + self.data[word]
            for st in All_stopWord :
                if word + st in self.data:
                    cnt = cnt + self.data[word+st]
            similar.append((i,cnt))
            similar.sort(key = lambda x:x[1], reverse = True)
        for i in similar:
            if cnt <= i[1]:
                print(i[0])
                cnt = i[1]
            else :
                break
            
    
w1 = SearchEngine('http://www.cnn.com', 'http://www.times.com', 'https://www.amazon.com') 
w2 = SearchEngine('http://www.cnn.com', 'http://www.times.com')
w3 = SearchEngine()

w1.addUrl('https://github.com','https://www.naver.com')
w3.addUrl('http://stackoverflow.com')

w1.removeUrl('http://www.cnn.com')
w2.removeUrl('http://stackoverflow.com')

w1.listUrls()
print()
w2.listUrls()
print()
w3.listUrls()
print()
print('사전 자료형 출력')
print(w1.getWordsFrequency())
print()
print('최다 출현 빈도 단어 출력')
for i in w1.getMaxFreqencyWords():
    print(i)
print()
print('최대 유사도 사이트 출력')
w1.searchUrlByWord("news")
```

### [8번 문제]
위 7번 문제에서 정의한 SearchEngine 클래스를 상속하여 SearchEngineWithOrderedWebWords 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

+ 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
+ getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.

  +  reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.
  +  reverse 인자에 true를 넣으면 오름차순으로 정렬한다.
> \>>> w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')  
>
> \>>> w4.getWordsFrequency()  
> [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]    
>  
> \>>> w4.getWordsFrequency(reverse=true)  
> [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]  

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    def __init__(self, *Url):
        for i in Url:
            self.append(i)
        l = []
        for i in self:
            if i != '':
                req = requests.get(i)
                source = req.text
                text =  self.extractText(source)  #순수 텍스트 추출
                text = self.stripSpecialChar(text) #특수문자 제거
                l += text.split()            #공백단위분리
        self.data =  self.wordOfFreq(l)
        
    def getWordsFrequency(self, reverse = False):
        if reverse:
            self.reverse = False
        else :
            self.reverse = True
        data = sorted( self.data.items(), key=lambda x:x[1], reverse=self.reverse)
        print(data)
w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
w4.getWordsFrequency()
print()
w4.getWordsFrequency(reverse=True)
```

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    def __init__(self, *Url):
        for i in Url:
            self.append(i)
        l = []
        self.c = 0
        self.s = 0
        self.iterCounter = 0
        for i in self:
            if i != '':
                req = requests.get(i)
                source = req.text
                text =  self.extractText(source)  #순수 텍스트 추출
                text = self.stripSpecialChar(text) #특수문자 제거
                l += text.split()            #공백단위분리
        self.data =  self.wordOfFreq(l)
        self.word = sorted( self.data.items(), key=lambda x:x[1], reverse= True)


    def getWordsFrequency(self, reverse = False):
        if reverse:
            self.reverse = False
        else :
            self.reverse = True
        data = sorted( self.data.items(), key=lambda x:x[1], reverse=self.reverse)
        print(data)
    def __iter__(self):
        self.c = 0
        if self.iterCounter == 0:
            self.s = len(self)
        else :
            self.s = len(self.word)
        return self
    def __next__(self):
        if self.iterCounter == 0:
            if self.c < self.s :
                r = self.c
                self.c = self.c + 1
                return self[r]
            else :
                self.iterCounter = self.iterCounter + 1
                raise StopIteration
        else :
            if self.c < self.s :
                r = self.c
                self.c = self.c + 1
                return self.word[r]
            else :
                self.iterCounter = self.iterCounter + 1
                raise StopIteration
w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
for i in w4:
    print(i)
```
